Visualization of the latent space using t-SNE

- t-Distributed stochastic neighbor embedding (t-SNE) minimizes the divergence between two distributions: a distribution that measures pairwise similarities of the input objects and a distribution that measures pairwise similarities of the corresponding low-dimensional points in the embedding

```python
from keras.layers import Input
from keras.models import Model
import numpy as np
import pandas as pd
import seaborn as sns
import cv2
from matplotlib import pyplot as plt
from sklearn.manifold import TSNE
from utils import *
```

```python
x_train = read_train_images()
x_val = read_validation_images()
```

```python
y_train = read_train_labels()
y_val = read_validation_labels()
```

```python
x_train_gray = np.array([cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) for img in x_train])
x_val_gray = np.array([cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) for img in x_val])
```

```python
print("Training set (images) shape: {shape}".format(shape=x_train.shape))
print("Validation set (images) shape: {shape}".format(shape=x_val.shape))
```

```python
print("Training set (images) shape: {shape}".format(shape=x_train_gray.shape))
print("Validation set (images) shape: {shape}".format(shape=x_val_gray.shape))
```

```python
enc = read_model('encoder_layer')
```

```python
enc.summary()
```

```python
encoder_layers = enc.layers[6]
inChannel = 3
x, y = 128, 128
num_classes = 2
input_img = Input(shape = (x, y, inChannel))
imageSize = 128
```

```python
def compute_TSNE_projection(X, encoder):
    # Compute latent space representation
    print("Computing latent space projection...")
    X_resh = np.reshape(X,(len(x_train_gray),128,128,1))
    X_encoded = encoder.predict(X_resh)
    print('X_encoded: '+ str(X_encoded.shape))
    
    # Compute t-SNE embedding of latent space
    print("Computing t-SNE embedding...")
    tsne = TSNE(n_components=2, init='pca', random_state=0)
    print('TSNE: ' + str(tsne.get_params()))
    
    X_encoded_resh = np.reshape(X_encoded,(len(x_train_gray),-1))
    print('X_encoded_reshaped: ' + str(X_encoded_resh.shape))
    
    # Returns embedding of the training data in low-dimensional space.
    X_tsne = tsne.fit_transform(X_encoded_resh)  
    return X_tsne
```

```python
X_tsne = compute_TSNE_projection(x_train_gray, enc)
```

```python
df = pd.DataFrame()
df['y'] = np.argmax(y_train, axis=1)
df['tsne-2d-one'] = X_tsne[:,0]
df['tsne-2d-two'] = X_tsne[:,1]
```

```python
df.head()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="y",
    palette=sns.color_palette("hls", 2),
    data=df,
    legend="full",
    alpha=0.3)
```
